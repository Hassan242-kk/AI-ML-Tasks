```python
# Hassan Khalid
# DHC - 2690
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset iris.csv
df = pd.read_csv('iris.csv')

# Basic info
print(df.shape)
print(df.columns)
print(df.head())

df.info()
print(df.describe())
```

```python
# Scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.show()

# Histogram
df.hist(figsize=(10,8))
plt.show()

# Boxplot
sns.boxplot(data=df)
plt.show()
```
