```python
# Import pipeline from transformers
from transformers import pipeline

# Load a lightweight pre-trained model (DistilGPT2)
# This model generates text based on input prompts
chatbot = pipeline("text-generation", model="distilgpt2")
```

```python
# Function to generate empathetic responses
def mental_health_bot(user_input):

    # Prompt engineering:
    # We guide the model to behave like a supportive assistant
    prompt = f"You are a kind and supportive mental health assistant.\nUser: {user_input}\nAI:"

    # Generate response from model
    response = chatbot(prompt, max_length=100, num_return_sequences=1)

    # Return generated text
    return response[0]['generated_text']
```

```python
# Test the chatbot with a sample input
print(mental_health_bot("I feel very stressed and anxious"))
```

```python
# Function with basic safety filtering
def safe_mental_bot(user_input):

    # List of sensitive words
    banned_words = ["suicide", "kill", "harm"]

    # Check if input contains harmful content
    if any(word in user_input.lower() for word in banned_words):
        return "I'm really sorry you're feeling this way. Please consider talking to a professional or someone you trust."

    # Otherwise generate response
    return mental_health_bot(user_input)

# Test safe version
print(safe_mental_bot("I feel lonely"))
```
