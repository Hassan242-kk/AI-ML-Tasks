```python
from transformers import pipeline
#Due to API quota limitations, a local transformer model (DistilGPT2) was used to implement the chatbot.
#Prompt engineering and safety filters were applied to ensure responsible responses.
chatbot = pipeline("text-generation", model="distilgpt2")

def health_chatbot(user_input):
    prompt = f"You are a helpful medical assistant. Give safe, general advice.\nUser: {user_input}\nAI:"

    response = chatbot(prompt, max_length=100)

    return response[0]['generated_text']

print(health_chatbot("What causes sore throat?"))
```

```python
#Safety handling in chatbot responses
def safe_chatbot(user_input):
    banned_words = ["suicide", "kill", "overdose"]

    if any(word in user_input.lower() for word in banned_words):
        return "Please seek professional help. I cannot assist with this."

    return health_chatbot(user_input)

print(safe_chatbot("Is paracetamol safe for children?"))
```
